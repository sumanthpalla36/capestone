```python
## Response time

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('Crime_Data_from_2020_to_Present.csv')

# Convert the 'Date Rptd' and 'DATE OCC' columns to datetime
df['Date Rptd'] = pd.to_datetime(df['Date Rptd'], errors='coerce')
df['DATE OCC'] = pd.to_datetime(df['DATE OCC'], errors='coerce')

# Calculate the response time in days
df['Response Time (days)'] = (df['Date Rptd'] - df['DATE OCC']).dt.days

# Display the first few rows of the dataframe with the new column
print("\nResponse time calculated. Here are the first few rows with the new column:")
print(df[['Date Rptd', 'DATE OCC', 'Response Time (days)']].head())

# Visualization of response time
plt.figure(figsize=(12, 6))
sns.histplot(df['Response Time (days)'], bins=30, kde=True)
plt.title('Distribution of Response Time (days)')
plt.xlabel('Response Time (days)')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(x=df['Response Time (days)'])
plt.title('Boxplot of Response Time (days)')
plt.xlabel('Response Time (days)')
plt.show()
```

```python
##Location and crime type


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, precision_score, f1_score
from sklearn.cluster import DBSCAN

# Load the dataset
df = pd.read_csv('Crime_Data_from_2020_to_Present.csv')

# Data Cleaning
# Drop irrelevant columns
df = df.drop(columns=['DR_NO', 'Date Rptd', 'Rpt Dist No', 'Mocodes', 'Cross Street'])

# Handle missing values
df = df.dropna()

# Convert date columns to datetime
df['DATE OCC'] = pd.to_datetime(df['DATE OCC'])
df['TIME OCC'] = df['TIME OCC'].apply(lambda x: f"{int(x):04d}")
df['TIME OCC'] = pd.to_datetime(df['TIME OCC'], format='%H%M').dt.time

# Extract temporal features
df['Hour'] = df['TIME OCC'].apply(lambda x: x.hour)
df['DayOfWeek'] = df['DATE OCC'].dt.dayofweek
df['Month'] = df['DATE OCC'].dt.month

# Encode categorical variables
df['Vict Sex'] = df['Vict Sex'].astype('category').cat.codes
df['Vict Descent'] = df['Vict Descent'].astype('category').cat.codes
df['Crm Cd Desc'] = df['Crm Cd Desc'].astype('category').cat.codes

# Define features and target for clustering
X_clustering = df[['LAT', 'LON']]

# Apply DBSCAN clustering algorithm to identify hotspots
dbscan = DBSCAN(eps=0.01, min_samples=5)
df['Cluster'] = dbscan.fit_predict(X_clustering)

# Visualize the clusters (hotspots)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='LON', y='LAT', hue='Cluster', palette='viridis')
plt.title('Crime Hotspots')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(title='Cluster', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Define features and target for predictive modeling
X = df[['Vict Sex', 'Vict Descent', 'LAT', 'LON', 'Hour', 'DayOfWeek', 'Month']]
y = df['Crm Cd Desc']

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train a Random Forest Classifier with hyperparameter tuning for better accuracy
model = RandomForestClassifier(n_estimators=200, max_depth=20, min_samples_split=5, min_samples_leaf=2, random_state=42)
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
print(confusion_matrix(y_test, y_pred))

# Check for labels with no predicted samples
unique_labels = set(y_test)
predicted_labels = set(y_pred)
missing_labels = unique_labels - predicted_labels

if missing_labels:
    print(f"Warning: No predicted samples for labels: {missing_labels}")

# Generate classification report with zero_division parameter
print(classification_report(y_test, y_pred, zero_division=1))

# Calculate precision and F1 score with zero_division parameter
precision = precision_score(y_test, y_pred, average='weighted', zero_division=1)
f1 = f1_score(y_test, y_pred, average='weighted', zero_division=1)

print(f'Precision: {precision}')
print(f'F1 Score: {f1}')
```
